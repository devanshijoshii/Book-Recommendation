# Most Popular Books

```python
most_popular=df.groupby(['Book-Title']).count()['Book-Rating'].reset_index()
```

```python
most_popular=most_popular[most_popular['Book-Rating']>550]
```

```python
most_popular=df.merge(most_popular,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Image-URL-M','Book-Rating_y','Book-Author']]
```

```python
most_popular=most_popular.sort_values('Book-Rating_y',ascending=False)
most_popular
```

# Collaborative Based

```python
user=df['User-ID'].value_counts()[:800].index
```

```python
final_df=df[df['User-ID'].isin(user)]
```

```python
book=final_df['Book-Title'].value_counts()[:800].index
```

```python
final_df=final_df[final_df['Book-Title'].isin(book)]
```

```python
final_df
```

```python
main_df=final_df.drop_duplicates('Book-Title')
```

```python
main_df
```

```python
pv=final_df.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating',fill_value=0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(pv)
```

```python
def recommendation(book_name):
    index=np.where(pv.index==book_name)[0][0]
    similar_book=sorted(list(enumerate(similarity[index])),key=lambda x:x[1],reverse=True)[1:6]
    for i in similar_book:
        print(pv.index[i[0]])
```

```python
book_named='1984'
```

```python
recommendation(book_named)
```
